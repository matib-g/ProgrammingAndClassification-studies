12. Implement a Cesar cipher using with a given shift parameter d in Python. Use functions translate() and maketrans()


```python
def cesar(text,d):
    alphabet = 'abcdefghijklmnopqrstuvwxyz'
    shifted = alphabet[d:]+alphabet[:d]
    return text.translate(text.maketrans(alphabet,shifted))
```

```python
d=5
c = cesar('hello',d)
print(c)
```

```python
cesar(c,26-d)
```

13. \*\*\* Find your favourite book in text format (if you cannot find it, try e.g. Heller’s
Catch 22).
(a) Remove all the punctuation.
(b) Remove all stop words.
(c) Change inflected words to their word stem (stemming words).
(d) List all the words that appear at least 100 times in the text. Than sort them in
alphabetical order.

```python
filename='01 - The Fellowship Of The Ring.txt'
with open(filename, 'r') as f:
    file_content = f.read() 
words = nltk.word_tokenize(file_content)
```

(a) Remove all the punctuation.

```python
filtred = [word.lower() for word in words if word.isalpha()]
filtred[0:20]
```

(b) Remove all stop words.

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
filtred = [word for word in filtred if word not in stop_words]
filtred
```

(c) Change inflected words to their word stem (stemming words).

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()

stem_words = [ps.stem(word) for word in filtred ]
stem_words
```

(d) List all the words that appear at least 100 times in the text. Than sort them in alphabetical order.

```python
from nltk.probability import FreqDist

f = FreqDist(stem_words)
at_least_100=sorted(list({k:v for (k,v) in f.items() if v >=100}))
at_least_100
```

15. \*\* Import all texts text1 -text9 from nltk.book. Check how many times the word
knight appears in text6 (Monty Python) and text7 (Wall Street Journal).

```python
from nltk.book import text1,text2,text3,text4,text5,text6,text7,text8,text9
```

```python
def word_count(text,word):
    count = 0
    for i in text:
        if(i==word):
            count+=1
    return count
```

```python
def word_count_stem(text,word):
    ps = PorterStemmer()
    count = 0
    for i in text:
        if(ps.stem(i)==word):
            count+=1
    return count
```

```python
word_count(text6,'knight')
```

```python
word_count_stem(text6,'knight')
```

```python
word_count(text7,'knight')
```

```python
word_count_stem(text7,'knight')
```

19. Generate a random number from a set L {1, 2, . . . , 6}. List all words from text6 with
exactly L letters.

```python
import random
```

```python
r = random.randint(1,6)
find_all_word(r)
```

```python
def find_all_word(length):
    text = remove_punctuation(text6)
    result = set()
    for word in text:
        if len(word)==length:
            result.add(word)
    return result
```
