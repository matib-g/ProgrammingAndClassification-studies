12. Implement a Cesar cipher using with a given shift parameter d in Python. Use functions translate() and maketrans()

```python
def cesar(ptext, d):
    alphabet = string.ascii_lowercase
    shifted_alphabet = alphabet[d:] + alphabet[:d]
    return ptext.translate(ptext.maketrans(alphabet, shifted_alphabet))
```

```python
d = 3
text = 'Implement a Cesar cipher using with a given shift parameter d in Python'
m = cesar(text, d)
print(m)
```

```python
cesar(m,26-d)
```

13. *** Find your favourite book in text format (if you cannot find it, try e.g. Heller’s
Catch 22).

```python
fName = 'Catch-22.txt'
with open(fName, 'r') as f:
    fileText = f.read()
```

(a) Remove all the punctuation.

```python
aName = 'Catch-22_a.txt'
a = open(aName, 'a')
aTrans = fileText.maketrans(dict.fromkeys(string.punctuation))
aWrite = fileText.translate(aTrans)

a.write(aWrite)
a.close()
aWrite[0:1000]
```

(b) Remove all stop words.

```python
from nltk.corpus import stopwords
#nltk.download('stopwords')

bName = 'Catch-22_b.txt'
b = open(bName, 'a')

stop_words = set(stopwords.words('english'))
words = nltk.word_tokenize(fileText)
filtered = [word for word in words if word not in stop_words]

tokensWithoutSW = ' '.join([str(elem) for elem in filtered])
b.write(tokensWithoutSW)
b.close()
tokensWithoutSW[0:1000]
```

(c) Change inflected words to their word stem (stemming words).

```python
#from nltk.stem.snowball import SnowballStemmer
from nltk.stem import PorterStemmer

cName = 'Catch-22_c.txt'
c = open(cName, 'a')

porter = PorterStemmer()

filtered = [porter.stem(word) for word in words]

stemWords = ' '.join([str(elem) for elem in filtered])
c.write(stemWords)
c.close()

stemWords[0:1000]
```

(d) List all the words that appear at least 100 times in the text. Than sort them in
alphabetical order.

```python
d = dict()
for word in filtered:
    if word in d:
        d[word] = d[word] + 1
    else:
        d[word] = 1

sort_d = dict(sorted(d.items(), key=lambda item: item[0]))
for key in list(sort_d.keys()):
    if sort_d[key] >= 100:
        print(key, ":", sort_d[key])
```

```python
def countWord(text,word):
    x = 0
    for i in text:
        if(i==word):
            x+=1
    return x

def countStemWord(text,word):
    ps = PorterStemmer()
    x = 0
    for i in text:
        if(ps.stem(i)==word):
            x+=1
    return x
```

19. Generate a random number from a set L {1, 2, . . . , 6}. List all words from text6 with
exactly L letters.

```python
import random

r = random.randint(1,6)

def llwords (l, t):
    text = removePunctuation(t)
    wordList = set()
    for word in text:
        if len(word) == l:
            wordList.add(word)
    return wordList

llwords(r, text6)
```
