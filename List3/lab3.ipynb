21.* For a given bitstring b list all bitstrings b’, such that the Hamming distance between
b and b’ is equal 1.

```python
def hamming_distance_equals_one(string):
    for i in range(len(string)):
        if(string[i]=='0'):
            yield string[0:i]+'1'+string[i+1:]
        else:
            yield string[0:i]+'0'+string[i+1:]
```

```python
def hamming_distance(string1, string2):
    dist_counter = 0
    if(len(string1)<len(string2)):
        sh = string1
        lo = string2
    else:
        sh = string2
        lo = string1
        
    for n in range(len(sh)):
        if sh[n] != lo[n]:
            dist_counter += 1
        dist_counter += len(lo)-len(sh)
    return dist_counter
```

22.* Construct a function that returns a Jaccard similarity for two sets. Beware that this
function needs to check if at least one of the sets is nonempty.

```python
def jaccard_similarity(s1,s2):
    inter = len(set(s1)&set(s2))
    union = len(set(s1)|set(s2))
    if(union!=0):
        return inter/union
    print("invalid arguments")
```

```python
s1 = 'abc'
s2 = 'abc'
jaccard_similarity(s1,s2)
```

23.* Construct a function that computes Jaccard similarity for two strings treated as bags
of words.

```python
def jaccard_similarity_strings_as_bags(s1,s2):
    b1 = string_to_bag(s1)
    b2 = string_to_bag(s2)
    inter = sum(bag_intersection(b1,b2).values())
    union = sum(bag_union(b1,b2).values())
    if(union!=0):
        return inter/union
    print("invalid arguments")
```

```python
def bag_intersection(b1,b2):
    res ={}
    for i in b1:
        if i in b2:
            res[i] = min(b1[i],b2[i])
        
    return res
    
```

```python
def bag_union(b1,b2):
    res ={}
    if(len(b1)>len(b2)):
        sm = b2
        lg = b1
    else:
        sm = b1
        lg = b2
        
    for i in lg:
        if i in sm:
            res[i] = max(lg[i],lg[i])
        else:
            res[i] = lg[i]
    return res
```

```python
def string_to_bag(s):
    res={}
    for i in s:
        if i in res:
            res[i]+=1
        else:
            res[i]=1
    return res
```

```python
jaccard_similarity_strings_as_bags("abc","abc")
```

30. **** For a given bitstring b and a natural number n list all bitstrings b’, such that the
Hamming distance between b and b’ is equal n.


```python
def hamming_distance_equals_n(string,n):
    combs = combinations(range(len(string)),n)
    for comb in combs:
        res = string
        for i in comb:
            res = replace_at_index(res,i)
        yield res
```

```python
def replace_at_index(string,i):
    if(string[i]=='0'):
        return string[0:i]+'1'+string[i+1:]
    else:
        return string[0:i]+'0'+string[i+1:]
        
```

```python
list(hamming_distance_equals_n('10101',4))
```

31. *** Construct a function that for a given string and a natural number k returns a set
of all its k-shingles.

```python
def k_shingles(string,k):
    for i in range(len(string)-k+1):
        res = ''
        for j in range(k):
            res+=string[i+j]
        yield res
```

```python
list(k_shingles('hello world',4))
```
